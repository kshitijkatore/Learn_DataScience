```python
## Feature Engineering

df['Date']=df['date'].str.split('-').str[0]
df['Month']=df['date'].str.split('-').str[1]
df['Year']=df['date'].str.split('-').str[2]
```

```python
df.head()
```

```python
## Converting Datatype
df['Date']=df['Date'].astype(int)
df['Month']=df['Date'].astype(int)
df['Year']=df['Date'].astype(int)
```

```python
df.info()
```

```python
## Drop date column
df.drop('date',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_hour']=df['arr_time'].str.split(':').str[0]
df['Arrival_minute']=df['arr_time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_minute']=df['Arrival_minute'].astype(int)
```

```python
df.info()
```

```python
df.drop('arr_time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Departure_hour']=df['dep_time'].str.split(':').str[0]
df['Departure_minute']=df['dep_time'].str.split(':').str[1]
```

```python
df['Departure_hour']=df['Departure_hour'].astype(int)
df['Departure_minute']=df['Departure_minute'].astype(int)
```
